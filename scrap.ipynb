# Sracping the Data

### Importing the libraries

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
import itertools
import csv
```

## Scraping For 2BHK property Details

```python
details=[]
price = []
price_per_unit = []
area=[]
facing=[]
status=[]
featues=[]
posted_by=[]
posted=[]
try:
    driver = webdriver.Chrome('C:\Program Files (x86)\chromedriver.exe')
    driver.get("https://www.propertiesguru.com/residential-search/2bhk-residential_apartment_flat-for-sale-in-new_delhi")
    cnt = 0
    check = {cnt:True}
    while check[cnt] == True:
        time.sleep(5)
        #parsing the page
        html = driver.page_source
        soup = BeautifulSoup(html , features='lxml')
        data = soup.find_all("div" , class_="filter-property-list detailurl")
        face = soup.find_all("div" , class_="row filter-pro-details")
        cnt = len(data)
        try:
            if check[cnt] == True:
                break
        except:
            check[cnt]=True     

        driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    driver.close()
    
    #extracting text_data from class->filter-property-list detailurl
    for a in data:
        details.append(a.find('h1', attrs={'class':'filter-pro-heading'}).text)
        price.append(a.find('span', attrs={'class':'price'}).text)
        price_per_unit.append(a.find('span', attrs={'class':'price-per-unit'}).text)
        area.append(a.find('div', attrs={'class':'col-4'}).text)
        status.append(a.find('div', attrs={'class':'col-5'}).text)
        featues.append(a.find('ul', attrs={'class':'pro-list'}).text)
        posted_by.append(a.find('span', attrs={'class':'owner-name'}).text)
        posted.append(a.find('span', attrs={'class':'owner-post'}).text)
    
    #extracting text-data from class->row filter-pro-details
    for c in face:
        facing.append(c.find('div', attrs={'class':'col-3'}).text)
except:
    pass
```

### Creating the DataFrame

```python
df = pd.DataFrame(columns = ['Details', 'Price', 'Price/unit', 'Area', 'Facing', 'Status', 'Features', 'Posted_By', 'Posted'])
df['Details'] = details
df['Price'] = price
df['Price/unit'] = price_per_unit
df['Area'] = area
df['Facing'] = facing
df['Status'] = status
df['Features'] = featues
df['Posted_By'] = posted_by
df['Posted'] = posted
```

```python
df.head()
```

```python
df.tail()
```

### Cleaning The Data

```python
for i in range(len(df)):
    df['Features'][i] = df['Features'][i][1:]
for i in range(len(df)):
    df['Price/unit'] = df['Price/unit'][i].replace("@", "")
    df['Posted'] = df['Posted'][i].replace("Posted:", "")
    df['Status'] = df['Status'][i].replace("Status", "")
    df['Facing'] = df['Facing'][i].replace("Facing", "")
    df['Area'] = df['Area'][i].replace("Area", "")
    df['Features'] = df['Features'][i].replace("\n", ", ")
```

```python
df.head()
```

```python
df.tail()
```

```python
# Creating csv file
df.to_csv('2-bhk.csv', index = False)
```
